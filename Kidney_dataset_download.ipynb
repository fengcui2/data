```python
dataset_dir = '/content/drive/MyDrive/Kidney_cancer_500_temp'
```

```python
import numpy as np
if not os.path.exists('/content/drive/MyDrive/Kidney_cancer_500'):
  os.makedirs('/content/drive/MyDrive/Kidney_cancer_500', exist_ok=True)
for data_type in ['train', 'val', 'test']:
  os.makedirs(os.path.join('/content/drive/MyDrive/Kidney_cancer_500', data_type), exist_ok=True)



```

```python
for cls in os.listdir(dataset_dir):
  os.makedirs(os.path.join('/content/drive/MyDrive/Kidney_cancer_500', 'train', cls), exist_ok=True)
  os.makedirs(os.path.join('/content/drive/MyDrive/Kidney_cancer_500', 'val', cls), exist_ok=True)
  os.makedirs(os.path.join('/content/drive/MyDrive/Kidney_cancer_500', 'test', cls), exist_ok=True)

  folder_imgs = os.listdir(os.path.join(dataset_dir, cls))

  for i in range(len(folder_imgs)):
    img = folder_imgs[i]
    rand = np.random.random()
    if rand < 0.1:
      shutil.copyfile(os.path.join(os.path.join(dataset_dir, cls, img)), os.path.join('/content/drive/MyDrive/Kidney_cancer_500', 'val', cls, img))
    elif rand < 0.2:
      shutil.copyfile(os.path.join(os.path.join(dataset_dir, cls, img)), os.path.join('/content/drive/MyDrive/Kidney_cancer_500', 'test', cls, img))
    else:
      shutil.copyfile(os.path.join(os.path.join(dataset_dir, cls, img)), os.path.join('/content/drive/MyDrive/Kidney_cancer_500', 'train', cls, img))
```
